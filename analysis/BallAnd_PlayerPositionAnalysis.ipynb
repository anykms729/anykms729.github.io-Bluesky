```python
import ast
import numpy as np
def calculate_distance_from_ball(player_pos,ball_pos):
  distance_list = []
  player_pos_list = [ast.literal_eval(s) for s in player_pos]
  ball_pos_list = [ast.literal_eval(s) for s in ball_pos]
  for pos in zip(player_pos_list, ball_pos_list):
    distance = np.sqrt((pos[0][0] - pos[1][0])**2 + (pos[0][1] - pos[1][1])**2)
    distance_list.append(distance)
  return distance_list
```

```python
for i in range(1,12):
  home_col = "homePlayers_"+str(i)+"_xyz"
  # print(home_col)
  df["homePlayers_"+str(i)+"_ballDistance"] = calculate_distance_from_ball(df[home_col], df["ball_xyz"])
  away_col = "awayPlayers_"+str(i)+"_xyz"
  # print(away_col)
  df["awayPlayers_"+str(i)+"_ballDistance"] = calculate_distance_from_ball(df[away_col], df["ball_xyz"])
```

```python
import json

# read the JSON file into a dictionary
with open('/content/drive/My Drive/MCI Women_s Files/g2312135_SecondSpectrum_meta.json', 'r') as f:
    player_profile_data = json.load(f)

# print the dictionary
home_player_profile_data = player_profile_data["homePlayers"]
away_player_profile_data = player_profile_data["awayPlayers"]
```

```python
home_player_profile_data
```

```python
home_player_pos = dict()
for player_profile in home_player_profile_data:
  home_player_pos[player_profile["ssiId"]] = player_profile["position"]
home_player_pos
```

```python
away_player_pos = dict()
for player_profile in away_player_profile_data:
  away_player_pos[player_profile["ssiId"]] = player_profile["position"]
away_player_pos
```

```python
for i in range(1,12):
  home_col = "homePlayers_"+str(i)+"_position"
  home_id = "homePlayers_"+str(i)+"_playerId"
  away_col = "awayPlayers_"+str(i)+"_position"
  away_id = "awayPlayers_"+str(i)+"_playerId"
  df[home_col] = df[home_id].apply(lambda x: home_player_pos[x])
  df[away_col] = df[away_id].apply(lambda x: away_player_pos[x])
```
